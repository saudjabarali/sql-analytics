```python
%%sql
create database crud_ex;
use crude-ex;
```

```python
%%sql
create table crud_ex.employees (
    emp_id int primary key,Name varchar(30),Department varchar (30), Salary int, join_date DATE);
```

```python
%%sql
#Q3. Fetch all employee details.
select * from crud_ex.employees;
```

```python
%%sql
#Q4. Display name and salary of employees earning more than 40,000.

select name,salary from crud_ex.employees where salary > 40000;
```

```python
%%sql
#Q5. Fetch employees who belong to the 'AI' department and joined after 2023.
select * from crud_ex.employees where Department= 'AI' and join_date> 2023;
```

```python
%%sql
#Q6. Display employees sorted by salary in descending order.
select * from crud_ex.employees order by salary desc;
```

```python
%%sql
#🟡 UPDATE – Practice

#Q7. Increase salary by 5,000 for employees in the 'AI' department.
update crud_ex.employees set salary= salary+5000 where Department ='AI';
```

```python
%%sql
#Q8. Update department to 'Data Science' for employee with emp_id = 101.
update crud_ex.employees set Department= 'Data Science' where emp_id=101;
```

```python
%%sql
#🔴 DELETE – Practice

#Q9. Delete employees earning less than 25,000.
delete from crud_ex.employees where salary <25000;
```

```python
%%sql
#Q10. Delete all records from the table without dropping the table.
truncate table crud_ex.employees;
```
