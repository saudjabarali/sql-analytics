```python
%%sql
#T9. AND
#Employees from AI department and salary > 50,000
select* from tcaw_ex.operator_employees where department ='AI' and salary > 50000;
```

```python
%%sql
#T10. OR
#Employees from AI or HR department
select * from tcaw_ex.operator_employees where department = 'AI' or department='HR';
```

```python
%%sql
#T11. IN

#Employees from departments: AI, Data, Finance
select * from tcaw_ex.operator_employees where department in ('AI', 'Data', 'Finance');
```

```python
%%sql
#T12. NOT IN

#Employees not belonging to HR
select * from tcaw_ex.operator_employees where department not in ('HR') ;
```
