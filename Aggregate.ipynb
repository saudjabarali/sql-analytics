```python
%%sql
#A1. COUNT-Count the total number of employees.
select count(*) as Total_number_of_employees from agg_ex.aggregate_employees;
```

```python
%%sql
#A2. SUM-Find the total salary paid to all employees.
select sum(salary) as Total_salary_paid from agg_ex.aggregate_employees;
```

```python
%%sql
#A3. AVG-Find the average salary of employees.
SELECT avg(salary) as Average_Salary from agg_ex.aggregate_employees;
```

```python
%%sql
#A4. MIN-Find the minimum salary in the company.
select min(salary) from agg_ex.aggregate_employees;
```

```python
%%sql
#A5. Max-Find the minimum salary in the company.
select max(salary) from agg_ex.aggregate_employees;
```

```python
%%sql
#a6-Find the number of employees in each department.
select department, count(*) as Total_emp_from_each_dept from agg_ex.aggregate_employees group by department;
```

```python
%%sql
#a7-Find the average salary per department.
select department , avg(salary) from agg_ex.aggregate_employees group  by department;
```

```python
%%sql
#A8.-Find the maximum salary in each department.
select department , max(salary) from agg_ex.aggregate_employees group  by department;
```

```python
%%sql
#A9.Find the total salary per city.
select city, sum(salary) from agg_ex.aggregate_employees group by city;
```

```python
%%sql
#A10.Find the average salary of employees from Bangalore only.
select city, avg(salary) from agg_ex.aggregate_employees where city='Bangalore';
```

```python
%%sql
#A11.Count employees department-wise, but only for employees earning more than 40,000.
select count(*) as employees_earn_more_thaan_40k from agg_ex.aggregate_employees where salary>40000;
```

```python
%%sql
#A12.Display departments having more than 2 employees.
select department, count(*) as emp_count from agg_ex.aggregate_employees group by department  having count(*)>2;
```

```python
%%sql
#A13.Display departments where average salary is greater than 50,000.
select department, avg(salary) from agg_ex.aggregate_employees group by department having avg(salary)>50000;
```

```python
%%sql
#A14.Find cities where total salary exceeds 1,00,000.
select city as Cities_exeeds_100000  from agg_ex.aggregate_employees group by city having sum(salary)>100000;
```
