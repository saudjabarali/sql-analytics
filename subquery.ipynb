```python
%%sql
#SQ4-Find employees who belong to departments that have more than 2 employees.
select * from subq_ex.subtb where department in
(select department from subq_ex.subtb group by department having count(*)>2 );
```

```python
%%sql
#SQ5-Find employees who work in Bangalore OR Mumbai (using a subquery, not direct WHERE).
select * from subq_ex.subtb where city in
(select city from subq_ex.subtb where city = 'Bangalore' or city= 'Mumbai');
```

```python
%%sql
#SQ6-Find employees whose salary is greater than the average salary of their department.
select * from subq_ex.subtb where 
(select department, avg(salary) as salary from subq_ex.subtb group by department ) ;
```

```python
%%sql
SELECT emp_id,
       emp_name,
       department,
       salary
FROM subq_ex.subtb e
WHERE salary >
      (
          SELECT AVG(salary)
          FROM subq_ex.subtb
          WHERE department = e.department
      );
```

```python
%%sql
#SQ7-Find departments where average salary is greater than overall average salary.
select department from subq_ex.subtb group by department having avg(salary) >
(select avg(salary) from subq_ex.subtb);
```

```python
%%sql
#SQ8-Display: emp_name salary overall average salary (same value for all rows)
select emp_name, salary , (select avg(salary) from subq_ex.subtb) as Avg_salary from subq_ex.subtb;
```

```python
%%sql
#sq9-From a subquery, get:department average salary Then select only departments where avg salary > 50,000.
select department from subq_ex.subtb group by department having avg(salary) > 50000;
```

```python
%%sql
SELECT department
FROM (
    SELECT department, AVG(salary) AS avg_salary
    FROM subq_ex.subtb
    GROUP BY department
) dept_avg
WHERE avg_salary > 50000;
```

```python
%%sql
#SQ10-Find employees who earn more than the average salary of their own department.
select * from subq_ex.subtb a where salary >
(select  avg(salary)  from subq_ex.subtb where department = a.department);
```

```python
%%sql
#SQ11=Find departments that have at least one employee earning more than 60,000.
select department from subq_ex.subtb group by department having max(salary) > 60000
```

```python
%%sql
select distinct department from subq_ex.subtb a
where EXISTS (
    select  emp_name from subq_ex.subtb b where a.department =b.department and salary> 60000);
```

```python
%%sql
select distinct department from subq_ex.subtb a
where NOT EXISTS (
    select  emp_name from subq_ex.subtb b where a.department =b.department and salary> 60000);
```
