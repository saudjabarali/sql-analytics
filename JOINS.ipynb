```python
%%sql
#j1
select e.emp_name, d.dept_name  from join_ex.employees e 
INNER JOIN join_ex.departments d 
ON e.dept_id = d.dept_id ;
```

```python
%%sql
select e.emp_name,d.dept_name from join_ex.employees e
LEFT JOIN join_ex.departments d on e.dept_id=d.dept_id;
```

```python

%%sql
select e.emp_name,d.dept_name from join_ex.employees e
RIGHT JOIN join_ex.departments d on e.dept_id=d.dept_id;
```

```python
%%sql
select e.emp_name, d.dept_name  from join_ex.employees e 
INNER JOIN join_ex.departments d 
ON e.dept_id = d.dept_id where dept_name='AI';
```

```python
%%sql
#j5-Find number of employees per department
select count(e.emp_name) as No_of_emp, d.dept_name from join_ex.employees e 
INNER JOIN join_ex.departments d on e.dept_id=d.dept_id group by d.dept_name;
```

```python
%%sql
#j6
select e.emp_name from join_ex.employees e 
LEFT JOIN join_ex.departments d on e.dept_id=d.dept_id where dept_name is null;
```

```python
%%sql
#🔥 J7 – JOIN + HAVING-Find departments having more than 1 employee
select d.dept_name as Dept_with_more_emps from join_ex.employees e 
LEFT JOIN join_ex.departments d on e.dept_id=d.dept_id group by dept_name having count(e.emp_name)>1
;
```

```python
%%sql
#j8-Show employees who work in the same department as Saud
select e2.emp_name, d.dept_name from join_ex.employees e1 
JOIN join_ex.employees e2 ON e1.dept_id=e2.dept_id 
JOIN join_ex.departments d ON e1.dept_id=d.dept_id
where e1.emp_name='Saud' and e2.emp_name <>'Saud';
```
