```python
%%sql 
#dt1
select current_timestamp as currenttime;
```

```python
%%sql
#DT2
select order_id, customer_name , amount, year(order_date) as year_of_order , order_date from dt_ex.orders_1;
```

```python
%%sql
#Find orders placed in 2024.
select * from dt_ex.orders_1 where year(order_date)=2024;
```

```python
%%sql
select * from dt_ex.orders_1 where month(order_date)=03;
```

```python
%%sql
#Find orders placed in last 1 year.
select * from dt_ex.orders_1 where order_date between current_timestamp - interval 1 year and current_timestamp;
```

```python
%%sql
#Find orders placed in last 1 year.
select * from dt_ex.orders_1 where order_date>=current_timestamp-interval 1 year;
```

```python
%%sql
#DT6
select order_id, customer_name , amount , order_date , datediff(current_date,order_date) as days_since_ordered from dt_ex.orders_1;
```

```python
%%sql

#Add 7 days to each order_date.
select order_id, customer_name , amount , order_date , order_date + interval 7 day as 7_days_from_order , datediff(order_date + interval 7 day, order_date) as difference from dt_ex.orders_1;
```

```python
%%sql
#dt9-Format order_date as: DD-MM-YYYY.
select order_id, customer_name , amount , order_date , date_format(order_date,'%a-%d-%m-%y') as dmy_format from dt_ex.orders_1;
```

```python
%%sql
#Find orders placed on weekend.
select * from dt_ex.orders_1 where dayofweek(order_date) in (1,7);
```

```python
%%sql
#🔥 DT11-:-Find month-wise total revenue.
select max(amount),month(order_date) as months from dt_ex.orders_1 group by month(order_date) order by months ;
```

```python
%%sql
#Find orders placed between two dates.
select  * from dt_ex.orders_1 where order_date between '2023-12-10' and '2025-12-10';
```

```python
%%sql
  select order_id,customer_name, amount, order_date,order_time,created_at, row_number() over(partition by month(order_date),year(order_date) order by order_date) as rn
    from dt_ex.orders_1
```

```python
%%sql
#Find first day of month from order_date.dt13
select dayname(order_date) , order_date from
(
    select order_id,customer_name, amount, order_date,order_time,created_at, row_number() over(partition by month(order_date) order by order_date) as rn
    from dt_ex.orders_1) as a
where rn=1;
```

```python
%%sql
#Find last day of month.
select order_date, last_Day(order_Date) as last_dayy from dt_ex.orders_1;
```

```python
%%sql
#Find customers who ordered today.
select * from dt_ex.orders_1
where order_date= current_date;
```
