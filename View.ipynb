```python
%%sql
#Create a view vw_dept_avg_salary that shows:department average salary per department
CREATE VIEW view_ex.deptavg as 
select department , avg(salary) as avg_salary from subq_ex.subtb group by department ;
```

```python
%%sql
select * from view_ex.deptavg;
```

```python
%%sql
create VIEW view_ex.hslevel as 
SELECT emp_name, salary,
    
case 
WHEN salary>50000 then 'High'
WHEN salary between 25000 and 49000 then 'Medium'
else 'Low'
end as Salary_level
from subq_ex.subtb ;
```

```python
%%sql 
select * from view_ex.hslevel;
```
