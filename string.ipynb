```python
%%sql
UPDATE str_ex.semp set username= 
LOWER(concat( substring(emp_name,1,3),emp_id));
```

```python
%%sql
#s5
select * from str_ex.semp where emp_name LIKE 'A%';
```

```python
%%sql
#s6
select * from str_ex.semp where emp_name LIKE '%khan%';
```

```python
%%sql
select instr(emp_name,' ') as firstnamelen from str_ex.semp ;
```

```python
%%sql
#s7
select emp_name, LEFT(emp_name,instr(emp_name,' ')) as FirstName from str_ex.semp;
```

```python
%%sql
#s8
select email, RIGHT(email,instr(reverse(email),'@')-1) as Domain from str_ex.semp;
```

```python
%%sql
#chatgpt answer
SELECT email,
       SUBSTRING_INDEX(email, '@', -1) AS domain
FROM str_ex.semp
WHERE email IS NOT NULL;
```

```python
%%sql
##s9
select city, LOWER(city) from str_ex.semp ;
```

```python
%%sql
#s10
select emp_name, COALESCE(email,'Not Available') as Email from str_ex.semp;
```

```python
%%sql
#s11
select * from str_ex.semp where length(email)>20;
```
