```python
%%sql
create database winf_ex;
```

```python
%%sql
create table winf_ex.ranktb as 
select * from subq_ex.subtb;
```

```python
%%sql
create table winf_ex.kumar as
select * from winf_ex.ranktb where salary > 50000;
```

```python
%%sql
#Example: Rank employees by salary in each department
SELECT emp_name,
       department,
       salary,
       ROW_NUMBER() OVER (
           PARTITION BY department
           ORDER BY salary desc
       ) AS row_num
FROM winf_ex.ranktb;
```

```python
%%sql
#Same rank for ties, skips numbers
SELECT emp_id, emp_name,
       department,
       salary,
       RANK() OVER (
           PARTITION BY department
           ORDER BY salary DESC
       ) AS rank_num
FROM subq_ex.subtb;
```

```python
%%sql
SELECT emp_name,
       department,
       salary,
       DENSE_RANK() OVER (
           PARTITION BY department
           ORDER BY salary DESC
       ) as dense_rank_
FROM winf_ex.ranktb;
```

```python
%%sql
#Average without collapsing rows
SELECT emp_name,
       department,
       salary,
       AVG(salary) OVER (PARTITION BY department) AS dept_avg_salary
FROM subq_ex.subtb;
```

```python
%%sql
#Running total
SELECT emp_name,
       salary,
       SUM(salary) OVER (
           ORDER BY emp_id
       ) AS running_total
FROM subq_ex.subtb;
```

```python
%%sql
#Salary difference from previous employee
SELECT emp_name,
       salary,
       salary - LAG(salary) OVER (ORDER BY emp_id) AS diff_from_prev
FROM subq_ex.subtb;
```

```python
%%sql
#Rank employees by salary within each department
select emp_name, department , salary, rank() 
over( partition by department order by salary desc) as rankbydept
from winf_ex.ranktb;
```

```python
%%sql
create view winf_ex.view_high1 as 
select emp_name,department,salary , rank()
over( partition by department order by salary desc ) as rank_highemp
from winf_ex.ranktb ;
```

```python
%%sql
create view winf_ex.high_sal_emp as
select emp_name	,department,	salary from winf_ex.view_high1 where rank_highemp=1;
```

```python
%%sql
#chatgpt answer
SELECT emp_id,
       emp_name,
       department,
       salary
FROM (
    SELECT emp_id,
           emp_name,
           department,
           salary,
           ROW_NUMBER() OVER (
               PARTITION BY department
               ORDER BY salary DESC
           ) AS rn
    FROM subq_ex.subtb
) t
WHERE rn = 1;
```

```python
%%sql
select emp_id,emp_name, department,salary, avg(salary)
  over(partition by department) as avg_sal from winf_ex.ranktb
```

```python
%%sql
#Show employees earning above department average (using window function)
select emp_id,emp_name, department, salary from
( select emp_id,emp_name, department,salary, avg(salary)
  over(partition by department) as avg_sal from winf_ex.ranktb) s
where salary>avg_sal;
```

```python
%%sql
#Calculate running total of salary
select emp_id,emp_name,department,salary, lag(salary) over() +salary as runningtotal from winf_ex.ranktb;
```

```python
%%sql
#Calculate running total of salary
select emp_id,emp_name,department,salary, sum(salary) over( order by emp_id)  as runningtotal from winf_ex.ranktb;
```

```python
%%sql
#Find salary difference from previous employee
select emp_name, department ,salary, salary - coalesce(lag(salary) over(order by emp_id),0)  as diff from winf_ex.ranktb;
```

```python
%%sql
#Assign row numbers ordered by salary
select emp_name,department , salary, row_number() over(order by salary desc) as rownum from winf_ex.ranktb;
```

```python
%%sql
#Top 2 earners in each department
select emp_id,emp_name,department,salary from 
(select emp_id,emp_name,department,salary, dense_rank() over(partition by department order by salary desc) as ranks 
from winf_ex.ranktb) f
where ranks in(1,2);
```
