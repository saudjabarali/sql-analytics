```python
%%sql
create database nulldb;
```

```python
%%sql
CREATE TABLE nulldb.nemp (
    emp_id INT PRIMARY KEY,
    emp_name VARCHAR(50),
    department VARCHAR(30),
    salary INT,
    bonus INT,
    city VARCHAR(30)
);
```

```python
%%sql
#26
select * from nulldb.nemp where salary is null;
```

```python
%%sql
#27
select * from nulldb.nemp where city is not null;
```

```python
%%sql
#28
select emp_name, coalesce(bonus,0)   as Bonus from nulldb.nemp;
```

```python
%%sql
#a29
select emp_name, COALESCE(salary,0)+ COALESCE(bonus,0) as Final_pay from nulldb.nemp ;
```

```python
%%sql
#A30-Count how many employees do not have bonus
select count(*) as " Emp who don't have bonus" from nulldb.nemp where bonus is null;
```

```python
%%sql
#a31
select avg(salary) from nulldb.nemp;
```
