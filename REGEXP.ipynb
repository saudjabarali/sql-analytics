```python
| Pattern  | Meaning         |
| -------- | --------------- |
| ^A       | Starts with A   |
| A$       | Ends with A     |
| [0-9]    | Any digit       |
| [A-Za-z] | Any letter      |
| .        | Any character   |
| .*       | Any sequence    |
| {n}      | Exactly n times |
| +        | One or more     |
| *        | Zero or more    |
```

```python
%%sql
#R1
select * from regex_ex.practice_users where name regexp '[0-9]';
```

```python
%%sql
#R2-Find invalid phone numbers.
select * from regex_ex.practice_users where phone not regexp '^[0-9]{10}$';
```

```python
%%sql
SELECT phone, LENGTH(phone)
FROM regex_ex.practice_users
WHERE user_id = 2;
```

```python
%%sql
SELECT 
    phone,
    phone REGEXP '^[0-9]{10}$' AS test_result
FROM regex_ex.practice_users;
```

```python
%%sql
select version();
```

```python
%%sql
SELECT 
    phone,
    LENGTH(phone) AS char_length,
    CHAR_LENGTH(phone) AS char_length2,
    HEX(phone) AS hex_value
FROM regex_ex.practice_users;
```

```python
%%sql
SELECT 
    phone,
    phone REGEXP '^[0-9]{10}$' AS is_valid
FROM regex_ex.practice_users;
```

```python
%%sql
SELECT 
    phone,
    REGEXP_LIKE(phone, '^[0-9]{10}$') AS is_valid
FROM regex_ex.practice_users;
```

```python
%%sql
SELECT 
    phone,
    phone REGEXP '^[0-9]\\{10\\}$' AS is_valid
FROM regex_ex.practice_users;
```

```python
%%sql
SELECT *
FROM regex_ex.practice_users
WHERE NOT REGEXP_LIKE(phone, '^[0-9]{10}$');
```

```python
%%sql
SELECT 
    REGEXP_LIKE('1234567890', '^[0-9]{10}$') AS test1,
    REGEXP_LIKE('12345abc10', '^[0-9]{10}$') AS test2;

```

```python
%%sql
#r3-Find emails not ending in .com.
select * from regex_ex.practice_users where email not regexp '[.][c][o][m]$'
```

```python
%%sql
select * from regex_ex.practice_users where email not regexp '.com$';
```

```python
%%sql
#r4-Find names with exactly 2 words.
SELECT *
FROM regex_ex.practice_users
WHERE REGEXP_LIKE(TRIM(name), '^[A-Za-z]+ [A-Za-z]+$');
```

```python
%%sql
#Extract username before @ from email.R5
select email , left(email, instr(email,'@')-1) as bef from regex_ex.practice_users ;
```

```python
%%sql
select email,  instr(email,'@') from regex_ex.practice_users;
```
