```python
%%sql
UPDATE case_ex.cemp SET Bonus=
CASE
WHEN department ='AI' THEN (salary*1.20)
WHEN department ='Data' THEN (salary*1.15)
ELSE (salary*1.10)
end;
```

```python
%%sql
#🔴 A22 – Experience Level (Date logic)
ALTER TABLE case_ex.cemp ADD COLUMN Experience_level varchar(20);
```

```python
%%sql
UPDATE case_ex.cemp SET Experience_level=
CASE
WHEN join_date between '2021-01-01' and '2023-12-31' THEN 'Intermed'
WHEN join_date<'2023-12-31' THEN 'Senior'

ELSE 'Fresher'
END;
```

```python
%%sql
#a23
SELECT *
FROM case_ex.cemp
WHERE
    CASE
        WHEN salary >= 60000 THEN 'High'
        WHEN salary BETWEEN 40000 AND 59999 THEN 'Medium'
        ELSE 'Low'
    END = 'High';
```
